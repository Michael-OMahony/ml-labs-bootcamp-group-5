## Environment_1

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X, y = exdata[predictors], exdata["Environment_1"]

rf1 = RandomForestClassifier()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
rf1.fit(X_train, y_train)
print((rf1.predict(X_test) == y_test).mean())
rf1.fit(X, y)
print((rf1.predict(X) == y).mean())
```

## Environment_2

```python
X, y = exdata[predictors], exdata["Environment_2"]

rf2 = RandomForestClassifier()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
rf2.fit(X_train, y_train)
print((rf2.predict(X_test) == y_test).mean())
rf2.fit(X, y)
print((rf2.predict(X) == y).mean())
```

```python
for dataset in [trainset, devset, testset]:
    env_1_probs = rf1.predict_proba(dataset[env_pred_feats])
    env_2_probs = rf2.predict_proba(dataset[env_pred_feats])
    for i in range(env_1_probs.shape[-1]):
        dataset[f"Environment_1:P{i}"] = env_1_probs[:, i]
    for i in range(env_2_probs.shape[-1]):
        dataset[f"Environment_2:P{i}"] = env_2_probs[:, i]
```
