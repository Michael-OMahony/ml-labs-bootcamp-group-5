```python
import numpy as np

np.arange(0, 10, 0.003)
```

```python
df = datapoint.bluetooth
step = df.Time.loc[1] - df.Time.loc[0]
time, rssi = [], []
for idx, t in df.iterrows():
    if t.Time > 0 and t.Time - df.Time.loc[idx-1] > 1:
        print(idx, t.Time)
        time.extend(np.linspace(df.Time.loc[idx-1], t.Time, 100))
        rssi.extend([0.]*100)
    else:
        time.append(t.Time)
        rssi.append(-t.Rssi)
```

```python
plt.figure(figsize=(17, 6))
plt.plot(time, rssi)
plt.xlabel('Time')
plt.ylabel('- RSSI')
```

```python
plt.figure(figsize=(17, 6))
plt.plot(time[:200], rssi[:200])
plt.xlabel('Time')
plt.ylabel('- RSSI')
```

```python
plt.figure(figsize=(17, 6))
plt.plot(time[270:450], rssi[270:450])
plt.xlabel('Time')
plt.ylabel('- RSSI')
```
